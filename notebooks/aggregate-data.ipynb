```python
from bikeshares.programs.nyc import CitiBike
from bikeshares.programs.chicago import Divvy
from bikeshares.programs.boston import Hubway
from bikeshares.program import TripSubset
import sys
import pandas as pd
import matplotlib as mpl
import mplstyle
```

```python
%matplotlib inline
```

```python
# NOTE: Restricting data to time common to all three programs
DATE_RANGE = ("2013-07-01", "2013-11-30")
```

```python
def get_gendered_trips(program, date_range=DATE_RANGE):
    trips = program.trips.between_times(*date_range)
    gender_constraint = trips["rider_gender"].notnull()
    gendered_subset = TripSubset(trips[gender_constraint])
    return gendered_subset
```

```python
def load_gendered(klass, trips_path):
    program = klass().load_trips(trips_path)
    trips = get_gendered_trips(program)
    return trips
```

```python
def get_female_pct(subset):
    return (subset["rider_gender"] == "F").mean() * 100
```

```python
citibike = load_gendered(CitiBike, "../raw-data/nyc-merged.csv")
```

```python
divvy = load_gendered(Divvy, "../raw-data/Divvy_Stations_Trips_2013/Divvy_Trips_2013.csv")
```

```python
hubway = load_gendered(Hubway, "../raw-data/hubway-updated-26-feb-2014/hubwaydata_10_12_to_11_13.csv")
```

```python
programs = pd.DataFrame([
    dict(name="Citi Bike", subset=citibike),
    dict(name="Divvy", subset=divvy),
    dict(name="Hubway", subset=hubway)
]).set_index("name")
```

## Female Ridership by Program

## Female Ridership by Day-of-Week

```python
from mplstyle.styles import simple
```

```python
mplstyle.set(simple)
mplstyle.set({
    "figure.figsize": (14, 8),
    "lines.linewidth": 0
})
BUZZRED = "#ee3322"
```

```python
weekdays = [ "Mon.", "Tues.", "Weds.", "Thurs.", "Fri.", "Sat.", "Sun." ]
```

```python
def gender_by_dow(subset):
    dow = subset["start_time"].apply(lambda x: x.dayofweek)
    grouped = subset.groupby(dow)
    by_dow = grouped.apply(get_female_pct)
    df = pd.DataFrame({
        "Female Ridership": by_dow,
        "Day": weekdays
    })
    return df.set_index("Day")
```

```python
def weekday_vs_weekend(subset):
    dow = subset["start_time"].apply(lambda x: x.dayofweek)
    is_weekday = dow < 5
    fpct = subset.groupby(is_weekday).apply(get_female_pct)
    return pd.Series({
        "Weekday": fpct[True],
        "Weekend": fpct[False],
        "diff": fpct[False] - fpct[True]
    })
```

```python
def plot_gender_by_dow(df, program_name):
    ax = df.plot(kind="bar", color=BUZZRED, legend=None)
    ax.set_xlabel("")
    ax.set_title("Female Ridership on {0} by Day of Week\n".format(program_name),
        fontsize="30",
        fontweight="bold")
    ax.set_ylabel("Trips Taken by Female Riders,\nas a Percentage of All Trips\n",
        fontsize="18")
    ax.xaxis.grid(False)
    ax.set_ylim(0, 30)
    ax.set_yticklabels([ "{0:.0f}%".format(y)
        for y in ax.get_yticks() ], fontsize="x-large")
    ax.set_xticklabels(weekdays, rotation=0,
        fontsize="18", fontweight="bold")
    return ax
```

### Citi Bike

```python
ax = plot_gender_by_dow(gender_by_dow(citibike), "Citi Bike")
```

```python
weekday_vs_weekend(citibike)
```

### Divvy

```python
ax = plot_gender_by_dow(gender_by_dow(divvy), "Divvy")
```

```python
weekday_vs_weekend(divvy)
```
